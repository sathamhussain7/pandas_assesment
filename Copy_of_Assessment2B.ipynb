###### Take each csv file , split that csv file into multiple categories (example csv files are added in the repo) 


###### consider if the codekata score exceeds 15000 points(present week) then make a csv on those observations as Exceeded expectations.csv


###### if  10000<codekata score<15000   (Reached_expectations.csv)



###### if  7000<codekata score<10000   (Needs_Improvement.csv)


###### if  codekate score < 7000        (Unsatisfactory.csv)

```python
#Reading the data from the college_1 data
df1 = pd.read_csv('college_1.csv')
#extracting the data based on codekatascore
expected = df1[df1['CodeKata Score']>15000]
#Reading the data from the college_2 data
df2 = pd.read_csv('college_2.csv')
expected2 = df2[df2['CodeKata Score']>15000]
#Concating the data extracted from both the files into a single Exceeded_expectation csv file
frames =[expected,expected2]
merge_df = pd.concat(frames)
merge_df.to_csv('exceeded_expectation.csv',index = True,header = True)
merge_df
```

```python
#Extracting the data from college1 and college2 files having codekata score with in the range of 10000-15000
re_expectn = df1[(df1['CodeKata Score']>10000) & (df1['CodeKata Score']<15000)]
re_expectn2 = df2[(df2['CodeKata Score']>10000) & (df2['CodeKata Score']<15000)]
#Concating the data extracted from both the files into a single reached expectation csv file
frames1 = [re_expectn,re_expectn2]
merge_df1 = pd.concat(frames1)
merge_df1.to_csv('reached_expectation.csv',index = True,header = True)
merge_df1
```

```python
need_imp = df1[(df1['CodeKata Score']>7000) & (df1['CodeKata Score']<10000)]
need_imp1 = df2[(df2['CodeKata Score']>7000) & (df2['CodeKata Score']<10000)]
#Concating the data extracted from both the files into a single needs improvement csv file
frames2 = [need_imp,need_imp1]
merge_df2 = pd.concat(frames2)
merge_df2.to_csv('needs_improvement.csv',index = True,header = True)
merge_df2
```

```python
un_sat = df1[(df1['CodeKata Score']<7000)]
un_sat1 = df2[(df2['CodeKata Score']<7000)]
#Concating the data extracted from both the files into a single un satisfactory csv file
frames3 = [un_sat,un_sat1]
merge_df3 = pd.concat(frames3)
merge_df3.to_csv('un_satisfactory.csv',index = True,header = True)
merge_df3
```
